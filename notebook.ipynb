```python
@file:DependsOn("org.knowm.xchart:xchart:3.5.2")

import org.knowm.xchart.QuickChart
import org.knowm.xchart.BitmapEncoder
import java.io.File

System.setProperty("java.awt.headless", "true")

val xData = doubleArrayOf(0.0, 1.0, 2.0)
val yData = doubleArrayOf(2.0, 1.0, 0.0)

val chartFile = File("chart_kotlin.png")
val chart = QuickChart.getChart("Sample Chart", "X", "Y", "y(x)", xData, yData)
BitmapEncoder.saveBitmap(chart, chartFile.absolutePath, BitmapEncoder.BitmapFormat.PNG)

println("Chart written to ${chartFile.absolutePath}")
println("Hello from IKotlin!")
println("2+2 = ${2+2}")
println("CWD: " + File(".").absolutePath)
println("Files: " + File(".").listFiles()?.joinToString { it.name })
```
